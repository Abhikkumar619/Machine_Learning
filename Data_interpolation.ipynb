## 1. Data Interpolation

### Linear Interpolation

```python
import numpy as np
```

```python
x=np.array([1,2,3,4,5])
y=np.array([2,4,6,8,10])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
```

```python
x_new=np.linspace(1,5,10) # create new x values
y_inter=np.interp(x_new, x, y)
```

```python
y_inter
```

```python
plt.scatter(x_new, y_inter)
```

## Cubic interpolation using scipy

```python
x=np.array([1,2,3,4,5,6])
y=np.array([1,8,27,64,125,216])
```

```python
plt.scatter(x,y)
```

```python
from scipy.interpolate import interp1d
```

```python
#create a cubic interpolation function.
f=interp1d(x,y,kind='cubic')
```

```python
f
```

```python
x_new=np.linspace(1,5,10)
```

```python
y_inter=f(x_new)
```

```python
plt.scatter(x_new, y_inter)
```

## polynomical interpolnation

```python
import numpy as np
x=np.array([1,2,3,4,5,6])
y=np.array([1,8,27,64,125,216])
```

```python
#interplate the data using polynomial interplation
p= np.polyfit(x,y,2) #Fit a 2nd degree polynomial to the data
```

```python
x_new=np.linspace(1,5,10) #create new x value.
y_interp=np.polyval(p, x_new) # interpalte y value
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x_new, y_interp)
```
