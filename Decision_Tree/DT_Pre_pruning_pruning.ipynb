## DEcision Tree classifier implementation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
dataset=load_iris()
```

```python
dataset.keys()
```

```python
print(dataset.DESCR)
```

```python
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
```

```python
df.head()
```

```python
dataset.target
```

```python
## independent and dependent features
x=df.iloc[:,:-1]
y=dataset.target
```

```python
x,y
```

```python
## train and test split
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.20, random_state=41)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
obj_tree=DecisionTreeClassifier()
```

```python
obj_tree.fit(x_train, y_train)
```

```python
x_train.head()
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(obj_tree, filled=True)
```

```python
## POST pruning with max_depth 2
obj_tree=DecisionTreeClassifier(max_depth=2)
```

```python
obj_tree.fit(x_train, y_train)
```

```python
plt.figure(figsize=(10,8))
tree.plot_tree(obj_tree, filled=True)
```

```python
## predication
y_pred=obj_tree.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
print(accuracy_score(y_pred, y_test))
```

```python
print(classification_report(y_pred, y_test))
```

## DecisionTree Prepruning And Hyperparameter

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
obj_tree1=DecisionTreeClassifier()
```

```python
parameter={
    'criterion':['gini','entropy', 'log_loss'],
    'splitter':['best', 'random'],
    'max_depth':[1,2,3,4,5],
    'max_features':['auto', 'sqrt', 'log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
clf=GridSearchCV(obj_tree1, param_grid=parameter, cv=5, scoring='accuracy')
```

```python
clf.fit(x_train, y_train)
```

```python
y_predict=clf.predict(x_test)
```

```python
y_predict
```

```python
clf.best_params_
```

```python
print(accuracy_score(y_predict, y_test))
```

```python
print(classification_report)
```
