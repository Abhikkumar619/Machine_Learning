SMOTE

```python
from sklearn.datasets import make_classification
import pandas as pd
```

```python
# x is independent features(input features)
#y is independent features(output features)
x,y=make_classification(n_samples=1000,
                    n_features=2,
                    n_redundant=0,
                    n_clusters_per_class=1,
                    weights=[0.90],
                    random_state=1
                    
                   )
```

```python
x
```

```python
y
```

```python
df1=pd.DataFrame(x, columns=['f1','f2'])
df2=pd.DataFrame(y, columns=['target'])
```

```python
df1.head()
```

```python
df2.head()
```

```python
final_df=pd.concat([df1,df2], axis=1)
```

```python
final_df.head()
```

```python
final_df['target'].value_counts()
```

```python
import matplotlib.pylab as plt
```

```python
plt.scatter(final_df['f1'], final_df['f2'], c=final_df['target'])
```

```python
from imblearn.over_sampling import SMOTE
```

```python
oversample=SMOTE() #insalize the smote.
x,y=oversample.fit_resample(final_df[['f1','f2']], final_df['target'])
```

```python
y.shape
```

```python
x.shape
```

```python
df1=pd.DataFrame(x, columns=['f1','f2'])
df2=pd.DataFrame(y, columns=['target'])
over_sampled=pd.concat([df1, df2], axis=1)
```

```python
over_sampled['target'].value_counts()
```

```python
plt.scatter(over_sampled['f1'], over_sampled['f2'], c=over_sampled['target'])
```
