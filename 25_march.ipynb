Q11. Are there any missing values in the dataset? Identify any missing values and describe how they may
impact your analysis.

```python
df1.isnull().sum()
```

```python

"""
Impact of missing value in data analysis. 
 
Biased Results: If missing values are not handled properly, it can introduce bias in the analysis. 
For example, if the missing values are not randomly distributed, it can lead to biased estimates and 
inaccurate conclusions.

Reduced Sample Size: Missing values reduce the effective sample size for analysis. This reduction in sample size can affect
                     the statistical power of the analysis and make it difficult to draw reliable conclusions or detect
                    meaningful patterns.

Distorted Relationships: Missing values can distort relationships between variables. For example, if there is a strong 
                         correlation between two variables, but one of them has many missing values, it can affect the
                         calculation of correlation coefficients and potentially mask the true relationship between the
                        variables.
"""
```
